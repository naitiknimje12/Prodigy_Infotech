```python
plt.figure(figsize=(12,6))
gt['Gender'].value_counts().plot(kind='bar',color=['green','pink'])
plt.grid(c='k',ls=':')
plt.xlabel('gender')
plt.ylabel('count')
plt.title('Genders Count',fontsize=25,fontweight='bold',c='k')
plt.text(0,500,'577',c='white')
plt.text(1,314,'87',c='k')
plt.show()
```
